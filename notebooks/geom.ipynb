```python
 import pandas as pd
 import pycedar
import gzip
```

```python
fitxy = pd.io.pickle.read_pickle('cache/geo.pck')
fitx = fitxy.pivot( 'x', 'y', 'fitx').interpolate().bfill().ffill()
fity = fitxy.pivot( 'x', 'y', 'fity').interpolate().bfill().ffill()
print( fitx )
```

```python
ga = pycedar.GeomAligner()
ga.xcont = ga.get_contours( fitx.interpolate() )
ga.ycont = ga.get_contours( fity.interpolate() )
```

```python
print( ga.xcont)
print( ga.ycont)
```

```python
ds42 = pycedar.load_templates_csv(
        gzip.open( 'data/align_csv/kaonp_1710_1500_42.csv.gz' ) ).loc[1500, :, : ] 

```

```python
prep = ga.prepare_data_sets( ds42.loc[(slice(200,300),slice(300,400)),:])
```

```python
print( prep)
```

```python
pnt = ga.compute_alignment( prep.loc[200, 350])
```
